```python
#Подключаем библиотеки
import numpy as np
import pandas as pd

#Читаем данные из файла
df = pd.read_excel("https://github.com/junaart/ForStudents/blob/master/Big_Data/Seminar_3/demo26.xlsx?raw=true", sheet_name = "ОПЖ")

#Отображаем данные
display(df)
```

```python
#Транспонируем таблицу
df = df.transpose()

#Переименовываем столбцы
df = df.rename(columns={0: "FIRST", 1: "SECOND", 2: "THIRD", 3: "FOURTH", 4: "FIFTH", 5: "SIXTH", 6: "SEVENTH", 7: "EIGHTH", 8: "NINTH"})

#Отображаем данные
display(df)
```

```python
#Удаляем строку содержащую названия столбцов
df.drop(labels = ['Год'], axis = 0, inplace = True)

#Отображаем данные
display(df)
```

```python
#Отобираем года, для которых значение во втором столбце больше среднего значения этого столбца
df[df['SECOND'] > df['SECOND'].mean()]
```

```python
#Выводим на экран все строки фрейма за четные года.
df.loc[::2,:]
```

```python
#Строим график. По оси X - года, по оси Y значения третьего столбца.
from matplotlib import pyplot as plt
plt.plot(df['THIRD'])
plt.grid(axis = 'both')
plt.show()
```

```python
#Добавляем к фрейму новый столбец категориальных данных по принципу: если значение третьего столбца в строке больше медианного значения этого столбца и значение четвертого столбца этой строки больше медианного значения этого столбца, то в новый столбец пишем 'good', в противном случае пишем 'bad'.
df = df.assign(decision = ((df.THIRD > df.THIRD.median())&(df.FOURTH > df.FOURTH.median())))
df.loc[df['decision'] == False, 'decision'] = 'bad'
df.loc[df['decision'] == True, 'decision'] = 'good'
display(df)
```

```python
#Сохраняем полученный фрейм с именем 'Test.csv'
df.to_csv('Test.csv')
```
