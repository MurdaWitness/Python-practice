```python
#Подключаем библиотеки
import numpy as np
import pandas as pd

#Читаем данные из файла
df=pd.read_csv("https://raw.githubusercontent.com/junaart/ForStudents/master/Big_Data/Seminar_4/data_1.csv",delimiter=",")

#Отображаем данные
display(df)
```

```python
#Cформировать два отдельных DataFrame из мужчин и женщин. Для женщин рассчитать и добавить в соответствующий DataFrame столбец возраста. Для мужчин рассчитать и добавить в соответствующий DataFrame столбец стажа.

#Преобразуем даты из строчного типа в тип datetime
df['Дата приема на работу'] = pd.to_datetime(df['Дата приема на работу'], format='%d.%m.%Y')
df['Дата рождения'] = pd.to_datetime(df['Дата рождения'], format='%d.%m.%Y')
#Задаём сегодняшний день
today = pd.Timestamp.today()

#Новый датафрейм в зависимости от пола
df_male = df[df['Пол'] == 'м']
#Создаём новый столбец "Стаж". Вычитаем из сегодняшнего дня столбец 'Дата приема на работу', и записываем в формате года
df_male = df_male.assign(Experience = (today - df_male['Дата приема на работу']).astype('<m8[Y]'))
display(df_male)
```

```python
#Новый датафрейм в зависимости от пола
df_female = df[df['Пол'] == 'ж']
#Создаём новый столбец "Возраст". Вычитаем из сегодняшнего дня столбец 'Дата рождения', и записываем в формате года
df_female = df_female.assign(Age = (today - df_female['Дата рождения']).astype('<m8[Y]'))
display(df_female)
```

```python
#Отобрать только сотрудников, имеющих детей и оклад, меньше среднего оклада сотрудников.
df[(df['Кол-во детей'] > 0) & (df['Оклад'] < df['Оклад'].mean())]
```

```python
#Выяснить, есть ли в списке юбиляры (кому в текущем году исполнилось количество лет, оканчивающееся на нуль или пять), получить этот список в виде DataFrame.
end_of_year = pd.Timestamp(year=2024, month=1, day=1)
df_jubilee = df
df_jubilee = df_jubilee.assign(Age = (end_of_year - df_jubilee['Дата рождения']).astype('<m8[Y]'))
df_jubilee = df_jubilee[(df_jubilee['Age'] % 5 == 0) | (df_jubilee['Age'] % 10 == 0)]
display(df_jubilee)
```

```python
#Получить DataFrame, в котором рассчитывается количество сотрудников по категориям: женат, замужем, холост, незамужем, разведен, разведена.
df['Семейное положение'] = df['Семейное положение'].astype('category')
ser_family = df['Семейное положение'].value_counts()
df_family = pd.DataFrame({'Семейное положение':ser_family.index, 'Количество':ser_family.values})
display(df_family)
```

```python
#Полученный DataFrame сохранить в формате csv
df_family.to_csv('Семейное_положение.csv')
```
